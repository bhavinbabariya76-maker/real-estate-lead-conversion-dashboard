```python
import pandas as pd
```

```python
df = pd.read_csv("real_estate_leads.csv")
```

```python
df.head()
```

```python
df["Conversion_Rate"] = (df["bookings"] /df["leads"]) * 100
```

```python
df["Cost_Per_Booking"] = (df["marketing_spend"] /df["bookings"])
```

```python
df.sort_values("Conversion_Rate",ascending=False)
```

```python
df.sort_values("Cost_Per_Booking")
```

```python
df.sort_values("bookings",ascending=False)
```
